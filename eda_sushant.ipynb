```python
import pandas as pd
import seaborn as sns
import numpy as np
import requests
import os
import json
import csv
import re
from dotenv import load_dotenv
load_dotenv()
```

```python
url = "https://chronicdata.cdc.gov/resource/us8e-ubyj.csv"
ASTHMA_API_TOKEN = os.getenv('ASTHMA_API_TOKEN')
headers = {
    "X-App-Token": ASTHMA_API_TOKEN,
}
 
response = requests.request("GET", url, headers=headers)
data = response.content

# f = open('asthma_data.csv', 'w')
# csv_file = csv.writer(f)
# for item in data:
#     csv_file.writerow(item)
# f.close()
```

```python
# Split the long string into a list of lines
data_split = data.decode('utf-8').splitlines()
data_split

# Open the file for writing
with open("asthma_data.csv", "w") as csv_file:
    # Create the writer object with tab delimiter
    writer = csv.writer(csv_file, delimiter = ',')
    for line in data_split:
        # Writerow() needs a list of data to be written, so split at all empty spaces in the line 
        line = re.split(',', line)
        line = [elem.replace('"', '') for elem in line]
        writer.writerow(line)
```
